```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("./datasets/homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel("area(sqr ft)")
plt.ylabel("price(US$)")
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

Equation for price: \
$price = 135.78767123 * area + 180616.43835616432$

```python
135.78767123 * 3300 + 180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

```python
d = pd.read_csv("./datasets/areas.csv")
d
```

```python
reg.predict(d)
```

```python
p = reg.predict(d)
d['prices'] = p
d
```

```python
d.to_csv("./datasets/prediction.csv", index=False)
```

### Exercise: 
Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020.

```python
canada_df = pd.read_csv('./datasets/canada_per_capita_income.csv')
canada_df
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(canada_df['year'], canada_df['per capita income (US$)'], color='red', marker='+')
plt.show()
```

```python
reg_model = linear_model.LinearRegression()
reg_model.fit(canada_df[['year']], canada_df[['per capita income (US$)']])
```

```python
reg_model.coef_
```

```python
reg_model.intercept_
```

$income = 828.46507522 * year -1632210.75785546$

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(canada_df['year'], canada_df['per capita income (US$)'], color='red', marker='+')
plt.plot(canada_df['year'], reg_model.predict(canada_df[['year']]), color='blue')
plt.show()
```

```python
# Prediction canada's per capita income in year 2020
reg_model.predict([[2020]])
```
