```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/lego_sets.csv')
df.head()
```

```python
theme = pd.read_csv('datasets/parent_themes.csv')
theme.head()
```

## 1. What percentage of all licensed sets ever released were Star Wars themed? 
Save your answer as a variable `the_force` in the form of on integer (e. g. 25).

```python
merged = df.merge(theme, left_on='parent_theme', right_on='name')
merged.head()
```

```python
merged.drop(columns='name_y', inplace=True)
merged.head()
```

```python
merged[merged['set_num'].isnull()]
```

```python
merged[merged['set_num'].isnull()].shape
```

```python
licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed.head()
```

```python
star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
star_wars.head(50)
```

```python
star_wars.shape[0]
```

```python
the_force = int(star_wars.shape[0] / licensed.shape[0] * 100)
print(the_force)
```

## 2. In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)?

Save you answer as a variable `new_era` in the form of an integer (e. g. 2012).

```python
licensed.head()
```

```python
licensed_sorted = licensed.sort_values('year')
licensed_sorted.head()
```

```python
licensed_sorted['count'] = 1

summed_df = licensed_sorted.groupby(['year', 'parent_theme']).sum().reset_index()
summed_df.head()
```

```python
summed_df.sort_values('count', ascending=False).drop_duplicates(['year'])
```

```python
max_df = summed_df.sort_values('count', ascending=False).drop_duplicates(['year'])
max_df.sort_values('year', inplace=True)
max_df
```

```python
new_area = 2017
```

## Break down number of sets by year

```python
clean_df = merged[~merged['set_num'].isnull()]
clean_df.head()
```

```python
clean_df['count'] = 1
sets_per_year = clean_df.groupby(['year']).sum().reset_index()[['year', 'count']]
sets_per_year.head(50)
```

```python
for index, row in sets_per_year.iterrows():
    print(row['year'], row['count'])
```

## Some visualizations

```python
import matplotlib.pyplot as plt
```

```python
theme
```

```python
theme['count'] = 1

plt.figure(figsize=(7,7))
theme.groupby(['is_licensed'])['count'].sum().plot(kind='pie', y='count', autopct='%1.2f%%')
plt.title('Licensed')
plt.axis('equal')
plt.ylabel("")
plt.show()
```

```python
df_cleaned = df.dropna(subset=['set_num'])
df_cleaned.isna().sum()
```

```python
plt.figure(figsize=(10,7))
df_cleaned.groupby(['parent_theme'])['set_num'].count().sort_values(ascending=False)[:10].plot(kind='bar')
plt.title('Top 10 parent themes')
plt.xlabel('parent theme')
plt.ylabel('counts')
plt.show()
```

```python
merged_df = df_cleaned.merge(theme, how="left", left_on='parent_theme', right_on='name')
merged_df
```

```python
merged_df.drop(columns=['name_y', 'count'], inplace=True)
merged_df.head()
```
